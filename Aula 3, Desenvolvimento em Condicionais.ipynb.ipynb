- Problema:
- Elabore o programa que faça a conversão de graus Fahrenheit (ºF) para
graus Celsius (ºC).
- Fórmula:    f = (c*1,8) + 32, c = 5*(f-32) / 9
- Passos para a resolução do problema (algoritmo):
Entrada de dados (leia)
Processamento (calcule)
Saída de dados (escreva)

Teste 1: Entrada: Fahrenheit = 32   Saída: Celsius = 0.0
Teste 2: Entrada: Fahrenheit = 60   Saída: Celsius = 15.555555555555557
Teste 3: Entrada: Fahrenheit = 55   Saída: Celsius = 12.77777777777779

```python
def start():
  print ("minha calculadora de temperatura.")
  temp = str(input('a temperatura a ser convertida e C ou F?'))
  if temp == 'F' or 'f':
    True
    tempf = float(input('digite a temperatura em F'))
    tempconvf = 5*(tempf-32)/9
    print(tempconvf)
    start()
  elif temp == 'C' or 'c':
    True
    tempc = float(input('digite a temperatura em C'))
    tempconvc = float(tempc*1.8)+32
    print(tempconvc)
    start()
  else:
    print('opcao invalida')
    start()
start()
```
