3. Construa o programa que tendo como dados de entrada dois pontos quaisquer do plano cartesiano, P(x1, y1) e Q(x2, y2), calcule a distância entre eles. Use a seguinte fórmula:
distância =  raiz (x_2-〖x_1)〗^2+〖(y_2-〖y_1〗_ )〗^2

```python
qx1 = float(input('digite a primeira cordenada do seu ponto Q: '))
qy1 = float(input('digite a segunda cordenada do seu ponto Q: '))
px2 = float(input('digite a primeira cordenada do seu ponto P: '))
py2 = float(input('digite a segunda cordenada do seu ponto P: '))

cx = (qx1 - px2)
cy = (qy1 - py2)
distancia = ((cx**2) + (cy**2))**0.5
print(distancia)
```

4. Desenvolva o programa que classifique dois valores inteiros quaisquer em ordem crescente. Os valores serão fornecidos pelo usuário.

```python
valor1 = int(input('digite seu primeiro valor'))
valor2 = int(input('digite seu segundo valor'))
if valor1 > valor2:
  print(f'o primeiro valor ({}) é maior que o segundo ({})'.format(valor1,valor2))
else:
  print(f'o segundo valor {} é maior que o primeiro {}'.format(valor2,valor1))
```

# Execicio 5

5. Construa o programa que calcule o peso ideal de uma pessoa.
  Utilize as seguintes fórmulas:
- Se homem, o peso ideal é calculado assim: (72,7 x altura) - 58;
- Se mulher, o peso ideal é calculado assim: (62,1 x altura) - 44,7.

Analise o problema e verifique quais são os dados que o usuário precisa fornecer (digitar).


```python

print('calculadora de peso ideal')
gender_definer = input('digite seu sexo (H ou M): ')
print(gender_definer)
if gender_definer.upper() == 'H':
    altura = float(input('(H selecionado) digite sua altura em M: '))
    pesoidealh = (72.7*altura)-58
    print(pesoidealh,'kg')
elif gender_definer.upper() == 'M':
    altura = float(input('(M selecionado) digite sua altura em M: '))
    pesoidealm = (62.1*altura)-44.7
    print(pesoidealm,'kg')
else:
  print('input invalido')
```
